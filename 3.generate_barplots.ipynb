```python
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# sns.set(style="darkgrid")
sns.set(style="whitegrid")
```

### Create DataFrame (Pandas) for all the results

```python
data = pd.read_csv('./old_results/common_without.txt')
impData = pd.read_csv('./old_results/common_with.txt')
col = ['Ratio Sum, $L$',
       '# Reagents, $k$',
       'Cases','CPU Time' , 
       'Avg. Completion Time, $T_{avg}$' , 
       'Avg. # Cells Used, $C_{avg}$' , 
       'Avg. # Valves Actuated, $V_{avg}$' , 
       'Avg. # Valve Actuations, $A_{avg}$']
n, k, cases, cpu, t, c, v, a = col
data.columns = col
impData.columns = col
```

```python
ntm = 'NTM'
hda = 'HDA + NTM'
data['Algorithm'] = ntm
impData['Algorithm'] = hda

# print (data.tail())
# print (impData.head())
```

```python
df = data.append(impData, ignore_index=True)
print (df.tail())
```

### Generate Barplots

```python
from NTM.utility import create_directory

dir_name = 'barplots'
create_directory(dir_name)

colormap = {hda : '#6aabef' , ntm : '#efab60'}
plt.rcParams["axes.labelsize"] = 20
padding = 0.1
```

```python
fig, axes = plt.subplots()
ax = sns.barplot(x=k, y=t, hue='Algorithm',
                 palette = colormap, 
                 data=df.loc[df[n] == 64])
ax.set_ylabel('Avg. Completion\n Time, $\mathbb{T}_{avg}$')
ax.tick_params(labelsize=14)

# to remove legend title
legend = axes.legend(frameon=True, shadow=True)
legend.texts[0].set_text(ntm)

plt.setp(ax.get_legend().get_texts(), fontsize='14')
plt.setp(ax.get_legend().get_texts(), fontstyle='italic')
plt.setp(ax.get_legend().get_title(), fontsize='10')

file_name = 'k_vs_t.png'
file_name = os.path.join(dir_name, file_name)
plt.savefig(file_name, pad_inches=padding, bbox_inches='tight')
```

```python
fig, axes = plt.subplots()
ax = sns.barplot(x=k, y=c, hue='Algorithm',
                 palette = colormap,
                 data=df.loc[df[n] == 64])
ax.tick_params(labelsize=14)
ax.set_ylabel('Avg. # Cells\n Used, $\mathbb{C}_{avg}$')
# to remove legend title
legend = axes.legend(frameon=True, shadow=True)
legend.texts[0].set_text(ntm)

plt.setp(ax.get_legend().get_texts(), fontsize='14')
plt.setp(ax.get_legend().get_texts(), fontstyle='italic')
plt.setp(ax.get_legend().get_title(), fontsize='10')

file_name = 'k_vs_c.png'
file_name = os.path.join(dir_name, file_name)
plt.savefig(file_name, pad_inches=padding, bbox_inches='tight')
```

```python
fig, axes = plt.subplots()
ax = sns.barplot(x=k, y=v, hue='Algorithm',
                 palette = colormap,
                 data=df.loc[df[n] == 64])
ax.tick_params(labelsize=14)
ax.set_ylabel('Avg. # Valves\n Actuated, $\mathbb{V}_{avg}$')

# to remove legend title
legend = axes.legend(frameon=True, shadow=True)
legend.texts[0].set_text(ntm)

plt.setp(ax.get_legend().get_texts(), fontsize='14')
plt.setp(ax.get_legend().get_texts(), fontstyle='italic')
plt.setp(ax.get_legend().get_title(), fontsize='10')

file_name = 'k_vs_v.png'
file_name = os.path.join(dir_name, file_name)
plt.savefig(file_name, pad_inches=padding, bbox_inches='tight')
```

```python
fig, axes = plt.subplots()
ax = sns.barplot(x=k, y=a, hue='Algorithm',
                 palette = colormap,
                 data=df.loc[df[n] == 64])
ax.tick_params(labelsize=14)
ax.set_ylabel('Avg. # Valve\n Actuations, $\mathbb{A}_{avg}$')

# to remove legend title
legend = axes.legend(frameon=True, shadow=True)
legend.texts[0].set_text(ntm)

plt.setp(ax.get_legend().get_texts(), fontsize='14')
plt.setp(ax.get_legend().get_texts(), fontstyle='italic')
plt.setp(ax.get_legend().get_title(), fontsize='10')

file_name = 'k_vs_a.png'
file_name = os.path.join(dir_name, file_name)
plt.savefig(file_name, pad_inches=padding, bbox_inches='tight')
```
